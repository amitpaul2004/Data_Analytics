```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r"D:\testing\BCT_3\data\student_fruit_dataset.xlsx")

print(data)
```

```python
fruit_count = data["Favorite_Fruit"].value_counts()

print(fruit_count)
```

```python
most_popular = data["Favorite_Fruit"].mode()[0]

print("Most Popular Fruit:", most_popular)
```

```python
percentage = data["Favorite_Fruit"].value_counts(normalize=True) * 100

print(percentage)
```

```python
fruit_count.plot(kind='bar')

plt.title("Favorite Fruit Popularity")
plt.xlabel("Fruit")
plt.ylabel("Number of Students")

plt.show()
```

```python
fruit_count.plot(kind='pie', autopct='%1.1f%%')

plt.title("Favorite Fruit Distribution")

plt.ylabel("")

plt.show()
```
